```python
import pandas as pd
import json
import seaborn as sns
import numpy as np

from sklearn.preprocessing import MinMaxScaler
```

```python
def visualise(filename: str):
    with open(filename) as f:
        data = [json.loads(line) for line in f]
    
    df = pd.DataFrame(data)
    # df.rename(columns={"score": "Score"}, inplace=True)
    df.groupby(["iteration", "length"]).mean()
    return sns.lineplot(data=df.groupby(["iteration", "length"]).mean(), x="iteration", y="score", hue="length", palette="Set2")
```

```python
sns.set(rc={'figure.figsize':(12,9)})

visualise("pipeline_data/pipeline_1/results.txt")
```

### Processed data

```python
def get_dataframes_from_txt_file(txt_file_path: str):
    train_fs, train_cs = [], []
    with open(txt_file_path) as f:
        for line in f:
            raw = json.loads(line)
            train_fs.append(pd.DataFrame(raw["features"]))
            train_cs.append(pd.DataFrame(raw["classes"]))

    X = pd.concat(train_fs)
    y = pd.concat(train_cs)

    return X, y

features_train, classes_train = get_dataframes_from_txt_file("pipeline_data/pipeline_1/processed_train.txt")
features_test, classes_test = get_dataframes_from_txt_file("pipeline_data/pipeline_1/processed_test.txt")
```

```python
from typing import List

def boxplot(df, column, user_ids: List[str]):
    sub_df = pd.DataFrame(
        data={
            "user_id": df.index,
            column: df[column].values,
        }
    )
    return sub_df, sns.boxplot(x="user_id", y=column, data=sub_df[sub_df["user_id"].isin(user_ids)])
```

```python
# features_train.Acc_X__standard_deviation
# ["10", "11", "12", "13"]

boxplot(features_train, "Acc_X__standard_deviation", ["10", "11", "12", "13"])
```
