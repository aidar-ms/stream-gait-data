```python
import json
import pandas as pd
import seaborn as sns
from typing import List


def boxplot(df, column, user_ids: List[str]):
    sub_df = pd.DataFrame(
        data={
            "user_id": df.index,
            column: df[column].values,
        }
    )
    return sub_df, sns.boxplot(x="user_id", y=column, data=sub_df[sub_df["user_id"].isin(user_ids)])


def visualise_density(filename: str):
    with open("pipeline_data/pipeline_2/cli/results/4.txt") as f:
        data = [json.loads(line) for line in f]

    df = pd.DataFrame(data)

    sns.set(rc={'figure.figsize': (12,9)})
    sns.displot(df, x="score", hue="length", kind="kde", palette="Set2")


def visualise(filename: str):
    with open(filename) as f:
        data = [json.loads(line) for line in f]
    
    df = pd.DataFrame(data)
    # df.rename(columns={"score": "Score"}, inplace=True)
    df.groupby(["iteration", "length"]).mean()

    sns.set(rc={'figure.figsize': (12,9)})

    return sns.lineplot(data=df.groupby(["iteration", "length"]).mean(), x="iteration", y="score", hue="length", palette="Set2")


def get_dataframes_from_txt_file(txt_file_path: str):
    train_fs, train_cs = [], []
    with open(txt_file_path) as f:
        for line in f:
            raw = json.loads(line)
            train_fs.append(pd.DataFrame(raw["features"]))
            train_cs.append(pd.DataFrame(raw["classes"]))

    X = pd.concat(train_fs)
    y = pd.concat(train_cs)

    return X, y
```
