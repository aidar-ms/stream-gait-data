```python
import json

import pandas as pd
import numpy as np
import seaborn as sns
```

```python

def get_features(filename: str):
    with open(filename) as f:
        features = [json.loads(line)["features"] for line in f]

    df = pd.concat([pd.DataFrame(f) for f in features])

    df.reset_index(inplace=True)

    df.rename({"index": "UserId"}, axis=1, inplace=True)

    return df
```

```python
features_8_test = get_features("pipeline_data/pipeline_2/cli/features/8_test.txt")

features_8_test[features_8_test["UserId"] == "1"]
```

```python
# PCA

user_1_features_8 = features_8_test[features_8_test["UserId"] == "1"].copy()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca.fit(user_1_features_8)

print(pca.explained_variance_ratio_)

print(pca.singular_values_)
```

```python
pd.DataFrame(pca.components_, columns=user_1_features_8.columns, index = ['PC1','PC2'])
```
