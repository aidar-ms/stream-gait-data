#### Participants didn't walk on one surface for the whole duration of a test and then move to another surface. Surfaces are almost in a random sequence
So let's try to get a portion of data which we'll have some variety of surface and to which we'll be able to apply TSfresh

```python
ts_data = data_sensor_0[(data_sensor_0["Surface"] < 5)].copy()
ts_data.insert(0, "Time", list(range(1, len(ts_data) + 1)))
ts_data
```

#### Let's try to extract some features with TSFresh

```python
from tsfresh import extract_features, select_features
from tsfresh.feature_selection.relevance import calculate_relevance_table
from tsfresh.feature_extraction.settings import MinimalFCParameters
from tsfresh.utilities.dataframe_functions import impute
```

```python
ts_columns = [c for c in ts_data.columns if c != "SensorLocation"]
ts_data[ts_columns]
```

```python

X = extract_features(ts_data[ts_columns], column_id="Surface", column_sort="Time", impute_function=impute, default_fc_parameters=MinimalFCParameters())
```

```python
X.dropna(axis=1, how="all", inplace=True)
```
