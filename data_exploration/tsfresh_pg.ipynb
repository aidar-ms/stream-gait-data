#### Let's try to extract some features with TSFresh

```python
from tsfresh import extract_features, select_features
from tsfresh.feature_selection.relevance import calculate_relevance_table
from tsfresh.feature_extraction.settings import MinimalFCParameters
from tsfresh.utilities.dataframe_functions import impute
```

```python
ts_columns = [c for c in data.columns if c != "SensorLocation"]
X = data[data["id"] < 800]
X.Surface
```

```python
X = extract_features(data[ts_columns], column_id="id", column_sort="id", impute_function=impute, default_fc_parameters=MinimalFCParameters())
```

```python
X
```

```python
X.dropna(axis=1, how="all", inplace=True)
```

```python
# Use binary array as a target
y = pd.Series(X.index.values == 1, X.index.values)
y
```

```python
X_filtered = select_features(X, y)
```
