```python
import pandas as pd
import json
import seaborn as sns
import numpy as np
```

```python
def visualise(filename: str):
    with open(filename) as f:
        data = [json.loads(line) for line in f]
    
    df = pd.DataFrame(data)
    # df.rename(columns={"score": "Score"}, inplace=True)
    df.groupby(["iteration", "length"]).mean()
    return sns.lineplot(data=df.groupby(["iteration", "length"]).mean(), x="iteration", y="score", hue="length")
```

### With __sum_values and __length columns

```python
sns.set(rc={'figure.figsize':(12,9)})

visualise("results.txt")
```

### Without __sum_values and __length columns

```python
visualise("results2.txt")
```
