```python
import pandas as pd
import json
import seaborn as sns
import numpy as np
```

```python
with open("results.txt") as f:
    data = [json.loads(line) for line in f]
```

```python
df = pd.DataFrame(data)
```

```python
df.rename(columns={"score": "Score"}, inplace=True)
```

```python
df.groupby(["iteration", "length"]).mean()
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.lineplot(data=df.groupby(["iteration", "length"]).mean(), x="iteration", y="Score", hue="length")
```
